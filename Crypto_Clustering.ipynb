```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "/Users/vrindapatel/Desktop/CryptoClustering/Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the csv file.
model = StandardScaler()
normalized_data = model.fit_transform(df_market_data)
df_market_data_normalized = pd.DataFrame(normalized_data, columns=df_market_data.columns, index=df_market_data.index)
df_market_data_normalized.head()
```

```python
# # Create a DataFrame with the scaled data
df_scaled = pd.DataFrame(normalized_data, columns=df_market_data.columns, index=df_market_data.index)

# Copy the crypto names from the original data
df_scaled['coin_id'] = df_market_data.index

# Set the coin_name column as index
df_scaled.set_index('coin_id', inplace=True)

# Display sample data
df_scaled.head()
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1, 12))
```

```python
# Create an empty list to store the inertia values
# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list

inertia = []
for i in k:
    model = KMeans(n_clusters=i, random_state=0)
    model.fit(df_scaled)
    inertia.append(model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve

elbow_data = {
    "K": k,
    "inertia": inertia
}

# create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)

# Display the DataFrame
df_elbow
```

```python


# Plot the Elbow curve
df_elbow.hvplot.line(
    x='K',
    y='inertia',
    title='Elbow Curve',
    xticks=k,
    width=800,
    height=400
)
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** Looking at the Elbow Curve graph, it appears that the elbow point occurs around (k = 4). This means that increasing the number of clusters beyond 4 results in diminishing returns in terms of reducing inertia (within-cluster sum of squares). Therefore, (k = 4) seems to be the best value based on this analysis.


---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialise the K-Means model using the best value for K
model = KMeans(n_clusters=4, random_state=0)
```

```python
# Fit the model to the data
model.fit(df_scaled)
```

```python
# Predict the clusters
predictions = model.predict(df_scaled)

# Print the result array of cluster value
print(predictions)
```

```python
# Create a copy of Dataframe
df_market_data_clustered = df_market_data.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
df_market_data_clustered["Predicted Cluster"] = model.labels_

# Display sample data
df_market_data_clustered.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Colour the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point
df_market_data_clustered.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by="Predicted Cluster",
    hover_cols=["coin_name"]
)
```

### Optimise Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3'.
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
market_data_pca = pca.fit_transform(df_scaled)

# Create a DataFrame with the principal component values
df_market_data_pca = pd.DataFrame(market_data_pca, columns=["PC1", "PC2", "PC3"])

# Display sample data
df_market_data_pca.head()
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
explained_variance = pca.explained_variance_ratio_
explained_variance
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** Based on the provided explained variance ratios, the total explained variance of the three principal components is approximately 89.5%.

**Total explained variance = explained_variance[0] + explained_variance[1] + explained_variance[2] = 0.3719856 + 0.34700813 + 0.17603793 = 0.89503166**

Therefore, the total explained variance of the three principal components is approximately 0.895 or 89.49%. This indicates that the three principal components capture a significant portion of the variability in the original data.

```python
# Create a new DataFrame with the PCA data.
# Creating a DataFrame with the PCA data
# Copy the crypto names from the original data
# Set the coinid column as index
# Display sample data

df_market_data_pca["coin_id"] = df_market_data.index
df_market_data_pca.set_index("coin_id", inplace=True)
df_market_data_pca.head()
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1, 12))
```

```python
# Create an empty list to store the inertia values


# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list
inertia = []
for i in k:
    model = KMeans(n_clusters=i, random_state=0)
    model.fit(df_market_data_pca)
    inertia.append(model.inertia_)
    
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {
    "K": k,
    "inertia": inertia
}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow_K = pd.DataFrame(elbow_data)

# Display the DataFrame
df_elbow_K
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for K
df_elbow_K.hvplot.line(
    x='K',
    y='inertia',
    title='Elbow Curve',
    xticks=k,
    width=800,
    height=400
)
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:**The Elbow curve suggests that around 4 clusters would be a reasonable choice for k. Beyond that point, the reduction in inertia becomes less significant.


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** It seems that the K-values remain consistent with the original data, but the inertia values are lower.

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialise the K-Means model using the best value for K
model = KMeans(n_clusters=4, random_state=0)
```

```python
# Fit the K-Means model using the PCA data
model.fit(df_market_data_pca)

# Predict the clusters
predictions = model.predict(df_market_data_pca)

# Print the result array of cluster value
print(predictions)
```

```python
# Create a copy of the DataFrame with the PCA data
df_market_data_clustered_PC = df_market_data_pca.copy()

# Add a new column to the DataFrame with the predicted clusters
df_market_data_clustered_PC["Predicted Cluster"] = model.labels_

# Display sample data
df_market_data_clustered_PC.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="PC1"` and `y="PC2"`. 
# Colour the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
df_market_data_clustered_PC.hvplot.scatter(
    x="PC1",
    y="PC2",
    by="Predicted Cluster",
    hover_cols=["coin_id"]
)
```

#### Answer the following question: 

  * **Question:** After visually analysing the cluster analysis results, what is the impact of using fewer features to cluster the data using K-Means?

  * **Answer:** 
*Reduced Dimensionality:
Reducing the number of features simplifies the clustering process by focusing on essential aspects of the data and eliminating noise.
It can lead to more efficient computations for high-dimensional data.

*Loss of Information:
Using fewer features may result in lost information, affecting clustering accuracy.
Relevant patterns or crucial structural information might be discarded.

*Increased Interpretability:
Fewer features make resulting clusters easier to interpret and understand.
Visualizing and explaining clusters becomes less challenging.

*Improved Scalability:
Clustering algorithms, like K-Means, can be computationally expensive for high-dimensional data.
Reducing features enhances scalability, allowing larger datasets to be processed efficiently.

*Feature Selection Sensitivity:
Carefully choosing relevant features is critical when using fewer dimensions.
Poor feature selection negatively impacts clustering results.
Remember that the impact of reduced dimensionality depends on the specific dataset and clustering goals.
